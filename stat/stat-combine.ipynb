```python
import pandas as pd
```

```python
df_order_local=pd.read_csv('order-local.csv', sep='\t')
df_order_router=pd.read_csv('order-router.csv', sep='\t')
```

```python
pd.merge(df_order_local, df_order_router, on='id', how='inner', suffixes=('_local', '_router'))
```
